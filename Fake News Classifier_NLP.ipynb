```python
df = df.dropna()
X = df.drop('label',axis=1)
y = df['label']
```

```python

messages = df.copy()
messages.reset_index(inplace=True)
```

```python
messages
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re 

ps = PorterStemmer()
corpus = []

for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
# Apply CountVectorizer
# Creating the Bag of words model 

from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(max_features=5000,ngram_range=(1,3))
X = cv.fit_transform(corpus).toarray()
```

```python
import tensorflow as tf 
```

```python
#Vocabulary size 
vocab_size = 5000
# This implies that we create 5000 words dictonary where every present in the text given is available , the no. of words should be less than or eq to  5000 
```

```python
from tensorflow.keras.layers import Embedding 
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential 
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense 
```

```python
onehot = [one_hot(words,vocab_size) for words in corpus]
onehot # This represents the specific index of the word according to the vocab_size we have arranged 
```

```python
sen_length = 20 
docs = pad_sequences(onehot,padding='pre',maxlen=sen_length)
print(docs)   # pre implies that we are padding at front, maxlen is the size of the word  
# Making the lenngth of the sentences equal by padding zeros to the one hot encoding  
```

```python
# Model 

embedding_features = 40 
model = Sequential()
model.add(Embedding(vocab_size,embedding_features,input_length=sen_length))
model.add(LSTM(100))  # 1 lstm layer of 100 neurons
model.add(Dense(1,activation='sigmoid'))
# We added dense layer because this is a classification problem 
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
X_final = np.array(docs)
Y_final = np.array(y)
```

```python
X_final.shape,Y_final.shape
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X_final,Y_final,random_state=42)
```

```python
# Model training 

model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred = model.predict(X_test)
```

```python
for i in range(len(y_pred)):
    if y_pred[i] >= 0.5 :
        y_pred[i] = 1 
    else : 
        y_pred[i] = 0
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,y_pred)
```
